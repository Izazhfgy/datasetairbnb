```python
import pandas as pd

df= pd.read_csv("latihan_1_penjualan_tiket.csv")
# Melihat 20 baris pertama dari dataset
df.head(20)
```

```python
# Melihat tipe data setiap kolom
df.dtypes
```

```python
# Mengubah kolom tanggal_film menjadi datetime
df['tanggal_film'] = pd.to_datetime(df['tanggal_film'], errors='coerce')
```

```python
# Cek jumlah duplikat
df.duplicated().sum()
df['is_duplicate_tanggal'] = df.duplicated(subset=['tanggal_film'], keep=False)

# Jika ingin melihat duplikatnya
df[df.duplicated()]
```

```python
# Menampilkan jumlah missing values per kolom
df.isnull().sum()
```

```python
print(df.describe())
```

```python
df['durasi_film'].fillna(df['durasi_film'].median(), inplace=True)
df['harga_tiket'] = df['harga_tiket'].fillna(0)
df['tiket_terjual'] = df['tiket_terjual'].fillna(0)
```

```python
# Menampilkan jumlah missing values per kolom
df.isnull().sum()
```

Pakai mean → saat datanya bersih dan simetris.

Pakai median → saat datanya banyak outlier atau distribusinya miring.

Pakai mode → saat data bersifat kategorik atau numerik dengan pengulangan.
