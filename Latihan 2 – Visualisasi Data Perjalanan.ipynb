```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("latihan_2_perjalanan.csv")
# Melihat tipe data setiap kolom
df.dtypes
```

```python
# Mengubah kolom tanggal_film menjadi datetime
df['tanggal_waktu'] = pd.to_datetime(df['tanggal_waktu'], errors='coerce')
```

```python
# Menampilkan jumlah missing values per kolom
df.isnull().sum()
```

```python
df.head(20)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python
df['tanggal'] = pd.to_datetime(df['tanggal_waktu']).dt.date
df['jam'] = pd.to_datetime(df['tanggal_waktu']).dt.hour
df.groupby('tanggal').size().plot(kind='line')  # volume order harian
sns.barplot(x='jam', y='harga', data=df)  # rata-rata harga per jam
```

```python
sns.scatterplot(x='jarak', y='durasi', data=df)
sns.scatterplot(x='harga', y='durasi', data=df)
```

```python
sns.boxplot(x='jam', y='customer_rating', data=df)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```
